```python
# loading libraries
import pandas as pd
import string
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
# importing data
tedtalk = pd.read_csv('transcripts.csv', encoding='utf-8')
```

```python
# remove escape character
ted = tedtalk['transcript'].str.replace("\\", "")   
```

```python
print(ted[0])                                   
```

```python
# contractions
import re
def decontracted(phrase):
    # specific
    phrase = re.sub(r"won\'t", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)

    # general
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase

# test = ted[0]
# print(decontracted(test))
ted_decon = ted.apply(decontracted)
type(ted_decon)
```

```python
# lowercase
ted_lower = ted_decon.str.lower()

# special characters
specials = '()—'
ted_spec = ted_lower.str.translate(str.maketrans(specials,' '*len(specials)))

# remove punctuation
ted_pun = ted_spec.str.translate(str.maketrans('','', string.punctuation))

# remove digits
ted_cleaned = ted_pun.str.translate(str.maketrans('','',string.digits))      
type(ted_cleaned)
```

```python
# tokenization
# tokenized_words = word_tokenize(ted_cleaned, 'english')
tokenized_words = ted_cleaned.apply(word_tokenize, 'english')

# remove stopwords
final_words = {}
for index, trans in enumerate(tokenized_words):
    final_words[index] = [word for word in trans if not word in stop_words] 
```

```python
final_words[0]
```

```python
# sentiment analysis
def sentiment_analyse(sentiment_text):
    score = SentimentIntensityAnalyzer().polarity_scores(sentiment_text)
    print("Overall sentiment dictionary is : ", score) 
    print("sentence was rated as ", score['neg']*100, "% Negative") 
    print("sentence was rated as ", score['neu']*100, "% Neutral") 
    print("sentence was rated as ", score['pos']*100, "% Positive")
    print("Sentence Overall Rated As", end = " ")
    if score['compound'] >= 0.05 : 
        print("Positive")   
    elif score['compound'] <= - 0.05 : 
        print("Negative") 
    else : 
        print("Neutral") 
```

```python
ted_cleaned.apply(sentiment_analyse)
```
